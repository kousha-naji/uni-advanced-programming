## Base Case

It is the condition that causes the callback to stop.
in the example you saw in the picture, the condition (n == 0) is the base case where the recursion stops.

## Recursive Case

A recursive case is a situation that calls again function with another input until it reaches a base case. in the example all situation that n>1, is recursive case.

## Q3

#### Recursive binary search

```python
def rec_binary_search(lst, target, low=0, high=None):

    sorter_lst = sorted(lst)

    if high is None:
        high = len(sorter_lst) - 1

    # base case
    if low > high:
        return -1
    mid = (low + high) // 2
    # recursive case
    if sorter_lst[mid] == target:
        return mid
    # recursive case
    elif sorter_lst[mid] > target:
        return rec_binary_search(sorter_lst, target, low, mid - 1)
    else:
        return rec_binary_search(sorter_lst, target, mid + 1, high)
```
